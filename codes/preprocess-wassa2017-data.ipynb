```python
import pandas as pd
```

```python
# root folder to access the raw wassa 2017 dataset
ROOT = './data/wassa2017_raw/'
```

## Import the raw dataset to do preprocessing

## Anger

```python
anger_train = pd.read_csv(f'{ROOT}anger-ratings-0to1.train.txt', sep='\t', header=None)
anger_train.head()
```

```python
anger_dev = pd.read_csv(f'{ROOT}anger-ratings-0to1.dev.gold.txt', sep='\t', header=None)
anger_dev.head()
```

```python
# label the header and drop the id column
# drop unnamed column
anger_train = anger_train.drop(labels=0, axis=1)
anger_train.head()
```

```python
# label the header and drop the id column
# drop unnamed column
anger_dev = anger_dev.drop(labels=0, axis=1)
anger_dev.head()
```

```python
# rename the column header for standardization
anger_train = anger_train.rename(columns={1: 'Text', 2: 'Label', 3:'Score'})
anger_train.head()
```

```python
# rename the column header for standardization
anger_dev = anger_dev.rename(columns={1: 'Text', 2: 'Label', 3:'Score'})
anger_dev.head()
```

```python
# export anger data for cleaning
anger_train.to_csv('./data/wassa_anger_train_raw.csv', index=False)
anger_dev.to_csv('./data/wassa_anger_dev_raw.csv', index=False)
```

## Fear

```python
fear_train = pd.read_csv(f'{ROOT}fear-ratings-0to1.train.txt', sep='\t', header=None)
fear_train.head()
```

```python
fear_test = pd.read_csv(f'{ROOT}fear-ratings-0to1.test.gold.txt', sep='\t', header=None)
fear_test.head()
```

```python
fear_dev = pd.read_csv(f'{ROOT}fear-ratings-0to1.dev.gold.txt', sep='\t', header=None)
fear_dev.head()
```

```python
# combine the train and the test set to become a larger train set
frames = [fear_train, fear_test]
fear_train_combined = pd.concat(frames)
fear_train_combined.head()
```

```python
# label the header and drop the id column
# drop unnamed column
fear_train_combined = fear_train_combined.drop(labels=0, axis=1)
fear_train_combined.head()
```

```python
# label the header and drop the id column
# drop unnamed column
fear_dev = fear_dev.drop(labels=0, axis=1)
fear_dev.head()
```

```python
# rename the column header for standardization
fear_train_combined = fear_train_combined.rename(columns={1: 'Text', 2: 'Label', 3:'Score'})
fear_train_combined.head()
```

```python
# rename the column header for standardization
fear_dev = fear_dev.rename(columns={1: 'Text', 2: 'Label', 3:'Score'})
fear_dev.head()
```

```python
# export anger data for cleaning
fear_train_combined.to_csv('./data/wassa_fear_train_raw.csv', index=False)
fear_dev.to_csv('./data/wassa_fear_dev_raw.csv', index=False)
```
