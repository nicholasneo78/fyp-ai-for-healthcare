```python
import pandas as pd
```

## Preprocess the short data
Retrieve those that only contains depressed data

```python
df_short = pd.read_csv('./data/tweets_combined.csv')
df_short.head()
```

```python
# drop unnamed column
df_short = df_short.drop(labels='Unnamed: 0', axis=1)
df_short
```

```python
# rename the column header for standardization
df_short = df_short.rename(columns={'tweet': 'Text', 'target': 'Label'})
df_short
```

```python
# extract short text data (mixed)
df_short.to_csv('./data/short_text.csv', index=False)
```

```python
# extract only the depressed data
df_short_dep = df_short[df_short['Label'] == 1]
df_short_dep
```

```python
# export the short depressed-only dataset
df_short_dep.to_csv('./data/short_text_only_depressed.csv', index=False)
```
