```python
# base path
DATA_PATH = './drive/MyDrive/fyp-code/codes/data/emotion_intensity/'
```

```python
# usual import
import pandas as pd
import numpy as np
```

```python
# import the short text data symbolic + subsymbolic method
short_data_sym_subsym = pd.read_csv(DATA_PATH+'emotion_intensity_depression_score_short_text_summarized.csv')[['text_raw', 'depression_score']]
short_data_sym_subsym.head(3)
```

```python
# import the short text data sentic API method
short_data_sentic = pd.read_csv(DATA_PATH+'emotion_intensity_sentic_depression_identification_short_data.csv')
short_data_sentic.head(3)
```

```python
# import the long text data symbolic + subsymbolic method
long_data_sym_subsym = pd.read_csv(DATA_PATH+'emotion_intensity_depression_score_long_text_summarized.csv')[['text_raw', 'depression_score']]
long_data_sym_subsym.head(3)
```

```python
# import the long text data sentic API method
long_data_sentic = pd.read_csv(DATA_PATH+'emotion_intensity_sentic_depression_identification_long_data.csv')
long_data_sentic.head(3)
```

### Merge the dataframe to get the depression scores

```python
# short data
short_data = pd.DataFrame()
short_data['text'] = short_data_sym_subsym.text_raw
short_data['score_sym_subsym'] = short_data_sym_subsym.depression_score
short_data['score_sentic'] = short_data_sentic.depression_score
short_data.head()
```

```python
# long data
long_data = pd.DataFrame()
long_data['text'] = long_data_sym_subsym.text_raw
long_data['score_sym_subsym'] = long_data_sym_subsym.depression_score
long_data['score_sentic'] = long_data_sentic.depression_score
long_data.head()
```

```python
# export the datasets
short_data.to_csv(DATA_PATH+'emotion_intensity_score_comparison_short.csv', index=False)
long_data.to_csv(DATA_PATH+'emotion_intensity_score_comparison_long.csv', index=False)
```
